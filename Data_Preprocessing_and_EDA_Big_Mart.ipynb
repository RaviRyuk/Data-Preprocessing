```python
train.drop('Outlet_Establishment_Year',inplace=True,axis=1)
```

```python
train.isna().sum()
```

```python
train['Item_Weight'] = train.groupby('Item_Type').Item_Weight.transform(lambda x : x.fillna(x.mean()))
```

```python
train.isna().sum()
```

```python
train['Item_Visibility'].value_counts()
```

```python
train.loc[train['Item_Visibility']==0,'Item_Visibility']=np.nan
```

```python
train.isna().sum()
```

```python
train['Item_Visibility']=train.groupby('Item_Identifier').Item_Visibility.transform(lambda x : x.fillna(x.median()))
```

```python
train.isna().sum()
```

```python
train['Outlet_Size'].value_counts()
```

```python
train['Outlet_Size'].fillna('Medium',inplace=True)
```

```python
train.isna().sum()
```

```python
train.loc[train['Outlet_Type'].str.startswith('Supermarket'),'Outlet_Type']='Supermarket'
```

```python
train.loc[train['Outlet_Type'].str.startswith('Grocery'),'Outlet_Type']='Grocery'
```
